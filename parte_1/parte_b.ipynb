```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import pandas as pd

```

```python
arquivo_nome = "dados_parte_b.txt"

# Abrindo e lendo o arquivo
with open(arquivo_nome, "r") as arquivo:
    conteudo = arquivo.read()

# Transformando o conteúdo em uma lista de floats
numeros = list(map(float, conteudo.split()))
tensoes_200 = pd.DataFrame(numeros)
resistencia_200 = (tensoes_200)*10/(5-tensoes_200) 

print(resistencia_200.mean())
```

```python
#### Agora, vamos calibrar o novo resistor





arquivo_nome = "dados_shunt_parte_b.txt"

# Abrindo e lendo o arquivo
with open(arquivo_nome, "r") as arquivo:
    conteudo = arquivo.read()

# Transformando o conteúdo em uma lista de floats
numeros = list(map(float, conteudo.split()))
numeros
```

```python
tensoes = np.array(numeros)
tensoes1 = 5 - tensoes
```

```python
dados = pd.DataFrame(tensoes1, columns=["A0"])
dados


## aqui se encontra as ddp causadas pelo resistor de 10 ohms
```

```python
# pela formula que encontramos, temos


resistencia_10_new = dados/(5-dados)*203
resistencia_10_new = dados/(5-dados)*203
print(resistencia_10_new)
print('A incerteza é: \n', np.sqrt((resistencia_10_new.std()**2/resistencia_10_new.count()) + 0.4**2))
print('A media é: \n', resistencia_10_new.mean())
## aqui esta a calibraçao do primeiro shunt
```
