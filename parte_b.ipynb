```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import pandas as pd

```

```python
arquivo_nome = "dados_parte_b.txt"

# Abrindo e lendo o arquivo
with open(arquivo_nome, "r") as arquivo:
    conteudo = arquivo.read()

# Transformando o conteúdo em uma lista de floats
numeros = list(map(float, conteudo.split()))
numeros
```

```python
tensoes = np.array(numeros)
tensoes = np.array([5 - tensoes, tensoes])
tensoes
```

```python
data = pd.DataFrame(tensoes.T, columns=['A0', "A1"])
data
```

```python
#
corrente = data["A1"]/200
```

```python
corrente = data["A0"]/10
corrente
```

```python
resistencia_10 = data["A0"]/corrente
resistencia_10.mean()
```
